```python
from matplotlib import pyplot as plt
from libsvm.svmutil import *
import random
```

```python
y_dirty, x_dirty = svm_read_problem('data.txt')

x, y, x1, x2, c = [], [], [], [], []
```

```python
for i, item in enumerate(x_dirty):
    value = list(item.values())
    if len(value) == 2:
        y.append(y_dirty[i])
        x.append(list(item.values()))
        x1.append(value[0])
        x2.append(value[1])

        if y_dirty[i] == 1:
            c.append('r')
        else:
            c.append('b')
```

```python
plt.scatter(x1, x2, c=c, s=10)
plt.show()
```

```python
combined = list(zip(x, y))
random.shuffle(combined)

x[:], y[:] = zip(*combined)
print(len(x))
```

```python
num_train = 600
m = svm_train(y[:num_train], x[:num_train], '-c 2 -t 1')
p_label, p_acc, p_val = svm_predict(y[num_train:], x[num_train:], m)
```

```python
c = []
x1 = [i[0] for i in x[num_train:]]
x2 = [i[1] for i in x[num_train:]]
for i in y[num_train:]:
    if i == 1:
        c.append('r')
    else:
        c.append('b')
plt.scatter(x1, x2, c=c, s=10)
plt.show()
```
